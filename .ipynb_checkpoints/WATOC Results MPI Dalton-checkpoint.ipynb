```python
def get_relative_error(data_dict,mols):
    mseries={}
    for mol in mols:
        df=data_dict[mol]
        rel_error_df=(((df.iloc[0:3]-df.loc['MRA'])/df.loc['MRA'])*100)
        mseries[mol]=rel_error_df
    return mseries
m=get_relative_error(data_dict_aug,mol_list2)
```

```python
from copy import deepcopy
import numpy as np
import yaml
import math


def incommon(*lists):
    """
    Return a set of elements common to all passed lists.

    Arbitrarily choose the first list of passed lists as starting point, 
    and then update based on the set.intersection_update result.

    :param lists: lists to be compared
    :return: set
    """
    s = set(lists[0])
    for l in lists[1:]:
        s.intersection_update(set(l))
    return s

def common_species():
    "Return a set of species common to the three main data sets"
    with open("data_sets/hg_data.yaml") as f, \
         open("data_sets/mw_fd.yaml") as g, \
         open("data_sets/gto_lr.yaml") as h:
        return incommon(yaml.load(f).keys(),
                        yaml.load(g).keys(),
                        yaml.load(h).keys())


def rms(l):
    "Return the root-mean-square of the passed list."
    return math.sqrt(average([x**2 for x in l]))

def myround(f, d=2):
    "Return str of passed float rounded to two decimals"
    return "%.{}f".format(d) % f

def isclose(a, b, tol=1e-4):
    "Evaluate whether two numbers are within the given threshold"
    if abs(a - b) <= tol:
        return True
    else:
        return False

def get_HG_species():
    "Return list of all species included in HG pol benchmark"
    with open("data_sets/hg_data.yaml") as f:
        return yaml.load(f).keys()
    
def normalize(l):
    "Subtract the average from each value, and return the normalized list"
    return [i - average(l) for i in l]
```

```python
# Set up figure
FS = 10
fig, ax = plt.subplots(figsize=(7, 5.25), dpi=300)

ax.set_ylabel("Relative Error [%]])", fontsize=FS)
ax.tick_params("x", labelsize=FS)
ax.tick_params("y", labelsize=FS)

y_range=3
ax.set_ylim((-y_range, y_range))

index=1

data_s=sData[index]
data_d=dData[index]

sdz=data_s['aug-cc-pVDZ'].values
stz=data_s['aug-cc-pVTZ'].values
sqz=data_s['aug-cc-pVQZ'].values

ddz=data_d['aug-cc-pVDZ'].values
dtz=data_d['aug-cc-pVTZ'].values
dqz=data_d['aug-cc-pVQZ'].values


# Collect useful information about each distribution
data = [
    {"re": sdz , "xlabel": "aug-cc-pVDZ", "n": len(sdz), "max": max(sdz), "mre": average([abs(re) for re in sdz])},
    {"re": stz, "xlabel": "aug-cc-pVTZ", "n": len(stz), "max": max(stz), "mre": average([abs(re) for re in stz])},
    {"re": stz, "xlabel": "aug-cc-pVQZ", "n": len(sqz), "max": max(sqz), "mre": average([abs(re) for re in sqz])},
    {"re": ddz , "xlabel": "d-aug-cc-pVDZ", "n": len(ddz), "max": max(sdz), "mre": average([abs(re) for re in ddz])},
    {"re": dtz, "xlabel": "d-aug-cc-pVTZ", "n": len(dtz), "max": max(dtz), "mre": average([abs(re) for re in dtz])},
    {"re": dtz, "xlabel": "d-aug-cc-pVQZ", "n": len(dqz), "max": max(dqz), "mre": average([abs(re) for re in dqz])},
]

pos = [i for i in range(len(data))]

# Make violin plot
vplots = ax.violinplot([d["re"] for d in data], 
                       positions=pos, 
                       showmeans=True,
                       showextrema=True,
                       points=data[0]["n"],
                       widths=0.9)

# Customize violin plot colors
for vplot in vplots["bodies"]:
    vplot.set_facecolor("skyblue")
    vplot.set_edgecolor("black")
    vplot.set_alpha(1)
    vplot.set_linewidth(2)

# Overlap data points with violin plot
for i, d in enumerate(data):
    ax.plot([i for _ in d["re"]], 
            d["re"], 
            linewidth=0.0,
            marker=".",
            color="black",
            markersize=10,
            markerfacecolor="red",
            markeredgecolor="black")
    
# Add x-axis labels
ax.set_xticklabels([""]+[d["xlabel"] for d in data])

# Add annotations for indicating sample sizes and average values
for i, d in enumerate(data):
    ax.annotate("n: {}".format(d["n"]),
               xy=(i, -1),
               fontsize=FS-2,
               ha="center",
               bbox=dict(boxstyle="round", fc="white", alpha=0.5))
    
    ax.annotate("MRE: {}".format(myround(d['mre'], d=2)),
               xy=(i, -1.7),
               fontsize=FS-2,
               ha="center",
               bbox=dict(boxstyle="round", fc="white", alpha=0.5))

plt.tight_layout()
```
