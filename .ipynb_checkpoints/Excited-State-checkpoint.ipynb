```python
from io import StringIO # Python 3 import
def createExcitedDaltonJSON(mol,xc,operator,basis):
    num_states = freq_json[mol][xc][operator]
    run_directory, dal_input, mol_input = dalton_excited_input(
            mol, xc, basis, num_states)
    outfile='/excited_'+'-'.join([mol,basis])+'.out'
    outfile=run_directory+outfile
    try:
        with open(outfile,'r')as daltonOutput:
            dj=daltonToJson()
            data=createExcitedJSON(dj.convert(daltonOutput),basis)
    except:
        print("did not find output file")
        print("Try and run molecule ",mol)
        d_out, d_error = run_dalton(run_directory, dal_input, mol_input)
        with open(outfile,'r')as daltonOutput:
            dj=daltonToJson()
            data=createExcitedJSON(dj.convert(daltonOutput),basis)
        pass
    return data
    
data=createExcitedDaltonJSON('Be','hf','excited-state','aug-cc-pVDZ')
freq_json = json.dumps(data, indent=4)
print(freq_json)

```
