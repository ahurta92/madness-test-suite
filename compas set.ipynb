```python
basis_list=['cc-pVDZ','cc-pVTZ']
```

```python

mols=[]
mol_list=glob.glob('molecules/*.mol')
for mol_i in mol_list:
    mols.append(mol_i.split('/')[1].split('.')[0])
    
print(mols)
    
```

```python
data={}

for mol in mols:
    data[mol]={}
    data[mol]['hf']={}
    data[mol]['hf']['excited-state']={}
    for basis in basis_list:
        moldata=dalton_reader.get_excited_json(mol,'hf',basis,False)
        if moldata is not None:
            data[mol]['hf']['excited-state'][basis]=moldata[basis]
        else:
            data[mol]['hf']['excited-state'][basis]=None
data            
```

```python
json_data=json.dumps(data,indent=4)
```

```python

with open('molecules/dalton-excited.json' , 'w') as f:
    f.write(json_data)
```
